```python
import pandas as pd
import xgboost as xgb
```

```python
df = pd.read_csv('data3.csv',sep=';')
```

```python
y = df.pop('uidb')
df.drop(['uida'],axis=1,inplace=True)
```

```python
def dum(df,name):
    dummies = df[name].str.get_dummies(sep=',').add_prefix(name+'_')
    df.drop([name],axis=1,inplace=True)
    dummies
    df = df.join(dummies)
    return df
# arr = ['education','experience','industry','interest','location','lookin4','offer','skill']
#arr = ['education']
arr = list(df)
for val in arr:
    df = dum(df,val)
df
```

```python
df.fillna(-1,inplace=True)
df.head(20)
```

```python
model = xgb.XGBClassifier(missing=9999999999)
```

```python
model.fit(df,y)
```

```python
model.predict(df.loc[10:22])
```
